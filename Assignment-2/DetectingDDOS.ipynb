```python
df[' Timestamp']=pd.to_datetime(df[' Timestamp'])
```

```python
df_resampled=df.set_index(' Timestamp').resample('6T').count()
```

```python
df_resampled
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_resampled.index, df_resampled[' Average Packet Size'], marker='o', linestyle='-', color='b')
plt.title('Incoming Packages')
plt.xlabel('Minutes')
plt.ylabel('Average Packet Size')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
